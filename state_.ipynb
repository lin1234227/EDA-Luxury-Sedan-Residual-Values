```python
def cat(curr):
    x = curr
    if x=='ILX' or x=='IS' or x=='A4' or x=='3' or x=='C':
        return 1
    elif x=='TLX' or x=='ES' or x=='GS' or x=='A6' or x=='5' or x=='E':
        return 2
    elif x=='RLX' or x=='LS' or x=='A8' or x=='7' or x=='S':
        return 3
    else:
        return 0
cars['model_level']=cars['model'].apply(lambda x: cat(x))
```
