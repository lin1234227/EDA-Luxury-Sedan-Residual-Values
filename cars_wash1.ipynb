```python
def find_year(curr_string):
    pattern = re.compile("\d\d\d\d")
    result = re.findall(pattern, curr_string)
    return result[0]
def find_capa(curr_string):
    pattern = re.compile("\d\.\d")
    result = re.findall(pattern, curr_string)
    if result:
        return result[0]
    else:
        return ""
def find_certified(curr_string):
    if "Certified" in curr_string:
        return 1
    else:
        return 0
```

```python
ILX['year'] = ILX["name"].apply(lambda x: find_year(x))
ILX['capacity'] = ILX['name'].apply(lambda x: find_capa(x))
ILX['Certified'] = ILX['name'].apply(lambda x: find_certified(x))
```

## powertrain

```python
def find_gearbox(curr_string):
    curr_string = str.lower(curr_string)
    if "auto" in curr_string and "manual" in curr_string:
        return "auto|manual"
    elif "auto" in curr_string or "a/t" in curr_string:
        return "auto"
    elif "manual" in curr_string:
        return "manual"
    elif "cvt" in curr_string:
        return "CVT"
    else:
        return ""
def find_dwheel(curr_string):
    pattern = re.compile(".WD")
    result = re.findall(pattern, curr_string)
    curr_string=str.lower(curr_string)
    if result:
        if result[0]=='4WD':
            return "AWD"
        return result[0]
    elif "4matic" in curr_string or "4X4" in curr_string:
        return "AWD"
    elif "all-wheel" in curr_string or "all wheel" in curr_string:
        return "AWD"
    elif "rear-wheel" in curr_string or "rear wheel" in curr_string:
        return "RWD"
    elif "2-wheel" in curr_string or "4X2" in curr_string:
        return "2WD"
    elif "front wheel" in curr_string:
        return "FWD"
    else:
        return ""
```

```python
ILX['gearbox'] = ILX['powertrain'].apply(lambda x: find_gearbox(x))
ILX['dwheel'] = ILX['powertrain'].apply(lambda x: find_dwheel(x))
```

## Looking

```python
def find_ext(curr_string):
    if "ext" in curr_string:
        if "|" in curr_string:
            result = curr_string.split("|")
            if "ext" in result[0]:
                return result[0].replace(" (ext) ", "")
            else:
                return result[1].replace(" (ext) ", "")
        else:
            pattern = re.compile("(.*?) \(ext\)")
            result = re.findall(pattern, curr_string)
            return result[0]
    else:
        return ""
def find_int(curr_string):
    if "int" in curr_string:
        if "|" in curr_string:
            result = curr_string.split("|")
            if "int" in result[0]:
                temp =  result[0].replace(" (int)", "")
                if temp[0]==" ":
                    return temp[1:]
            else:
                temp =  result[1].replace(" (int)", "")
                if temp[0]==" ":
                    return temp[1:]
        else:
            pattern = re.compile("(.*?) \(int\)")
            result = re.findall(pattern, curr_string)
            if len(result)==0:
                return ""
            temp=result[0]
            if temp[0]==" ":
                    return temp[1:]
    else:
        return ""
```
